# Assignment 1: Temperature Check


```python
temperature = int(input("Please current temperatur: "))

if temperature <10:
    print("It's too cold")

elif temperature >=10 and temperature <=25:
    print("It's cool outside!")

else:
    print("It's hot outside!")
```

 # Assignment 2: ATM Withdrawal System

```python
withdraw = int(input("Enter the withdraw amount :"))

total_balance = 50000

if withdraw > total_balance:
    print("Insufficient Balance")

elif withdraw <= total_balance:
    print("Transaction Successful")
    remain_balance = total_balance - withdraw
    print("Remaining Balance",remain_balance)

else:
    print("Invalid")
```

 # Assignment 3: Student Grade Calculator 

```python
marks = int(input("Enter your marks: "))

if marks >= 90:
    print("A+")
elif marks >=80:
    print("A")

elif marks >= 70:
    print("B")
elif marks >= 60:
    print("C")
else:
    print("Fail") 
```

# Assignment 4: Bus Ticket Discount 

```python
age = int(input("Please Enter your age: "))

if age < 5:
    print("Free")

elif age <= 18:
    print("50% Discount")
elif age >=60:
    print("30% Discount")
else:
    print("Full Price")
```

# Assignment 5: Restaurant Bill with Discount

```python
total_bill = int(input("Enter your total bill: "))

if total_bill > 1000:

    discount = total_bill *10/100
    final_bill_amount = total_bill - discount
    print("Final bill amount",final_bill_amount)

else: 
    print("No discount")
    print("Final bill amount: ",total_bill)
```

# Assignment 7: Traffic Light System 

```python

color = input("Enter a color (red, yellow or green): ").lower()

red = "red"
yellow = "yellow"
green = "green"

if color == red:
    print("Stop")

elif color == yellow:
    print("Ready to go")

elif color == green:
    print("Go")
else:
    print("Invalid color")
```

 # Assignment 8: Leap Year Checker

```python
year = int(input("Enter year: "))

if year % 400 == 0:
    print("Leap Year")
elif year % 4 ==0 and year % 100 !=0: 
    print("Leap Year")

else:
    print("Not a Leap Year")
```
